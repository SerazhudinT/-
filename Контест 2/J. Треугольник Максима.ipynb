```python
N = int(input())
f = float(input())

a = 30
b = 4000

for i in range(N - 1):
    newf, word = input().split()
    newf = float(newf)
    
    eps = abs(newf - f) / 2
    
    if word == 'closer':
        if newf > f:
            a = max(a, newf - eps)
        elif newf < f:
            b = min(b, newf + eps)
    elif word == 'further':
        if newf < f:
            a = max(a, f - eps)
        elif newf > f:
            b = min(b, f + eps)
            
    f = newf

print(a, b)
```
