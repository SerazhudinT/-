```python
ans = 'CONSTANT'
flag = True

lastElem, elem = int(input()), int(input())
end = -2 * 10**9

while elem != end:
    if flag:
        if lastElem < elem:
            ans = 'ASCENDING'
        elif lastElem > elem:
            ans = 'DESCENDING'
            
        flag = False
    else:
        if lastElem < elem:
            if ans == 'CONSTANT':
                ans = 'WEAKLY ASCENDING'
            elif ans in ('DESCENDING', 'WEAKLY DESCENDING'):
                ans = 'RANDOM'
        elif lastElem > elem:
            if ans == 'CONSTANT':
                ans = 'WEAKLY DESCENDING'
            elif ans in ('ASCENDING', 'WEAKLY ASCENDING'):
                ans = 'RANDOM'
        elif lastElem == elem and ans in ('ASCENDING', 'DESCENDING'):
            ans = 'WEAKLY ' + ans

    lastElem, elem =  elem, int(input())

print(ans)
```
