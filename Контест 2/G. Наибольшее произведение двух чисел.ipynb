```python
arr = list(map(int, input().split()))

max1, max2 = max(arr[:2]), min(arr[:2])
min1, min2 = max2, max1

for i in range(2, len(arr)):
    if max1 < arr[i]:
        max1, max2 = arr[i], max1
    elif max2 < arr[i]:
        max2 = arr[i]
    
    if min1 > arr[i]:
        min1, min2 = arr[i], min1
    elif min2 > arr[i]:
        min2 = arr[i]

if max1 * max2 > min1 * min2:
    print(max2, max1)
else:
    print(min1, min2)
```
