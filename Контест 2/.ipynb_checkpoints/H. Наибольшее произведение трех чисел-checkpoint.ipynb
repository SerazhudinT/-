```python
arr = list(map(int, input().split()))

max1, max2, max3 = sorted(arr[:3], reverse=True)
min1, min2 = max3, max2

for i in range(3, len(arr)):
    if max1 < arr[i]:
        max1, max2, max3 = arr[i], max1, max2
    elif max2 < arr[i]:
        max2, max3 = arr[i], max2
    elif max3 < arr[i]:
        max3 = arr[i]
    
    if min1 > arr[i]:
        min1, min2 = arr[i], min1
    elif min2 > arr[i]:
        min2 = arr[i]

if max1 * max2 * max3 > max1 * min1 * min2:
    print(max3, max2, max1)
else:
    print(min1, min2, max1)
```
