```python
def nearMine(arr, i, j, N, M):
    for k in range(-1, 2):
        for t in range(-1, 2):
            if 0 <= i + k < N and 0 <= j + t < M:
                if arr[i + k][j + t] == '*':
                    arr[i][j] += 1
        
    
N, M, K = map(int, input().split())
       
arr = [M * [0] for i in range(N)]

for i in range(K):
    p, q = map(int, input().split())
    
    arr[p - 1][q - 1] = '*'
    
for i in range(N):
    for j in range(M):
        if arr[i][j] != '*':
            nearMine(arr, i, j, N, M)

        print(arr[i][j], end=' ')
    print()
```
