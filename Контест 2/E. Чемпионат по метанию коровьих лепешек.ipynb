```python
def isVasya(arr, maxIndex, i):
    return arr[i] % 10 == 5 and maxIndex < i and arr[i] > arr[i + 1]

def VasyaPlace(arr, VasyaValue):
    place = 1
    
    for i in range(len(arr)):
        if arr[i] > VasyaValue:
            place += 1
            
    return place

N = int(input())
arr = list(map(int, input().split()))

maxValue, maxIndex = arr[0], 0
maxVasya = 0

for i in range(N):
    if maxValue < arr[i]:
        maxValue, maxIndex = arr[i], i

for i in range(1, N - 1):
    if isVasya(arr, maxIndex, i):
        maxVasya = max(maxVasya, arr[i])

if maxVasya == 0:
    place = 0
else:
    place = VasyaPlace(arr, maxVasya)

print(place)
```
