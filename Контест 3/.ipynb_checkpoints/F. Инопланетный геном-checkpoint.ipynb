```python
with open('input.txt') as fin:
    s1, s2 = fin.read().split()
    
    dict1 = {}
    for i in range(len(s1) - 1):
        dict1[s1[i:i + 2]] = dict1.get(s1[i:i + 2], 0) + 1
    
    set2 = set()
    for i in range(len(s2) - 1):
        set2.add(s2[i:i + 2])
        
    number = 0
    for elem in set2:
        number += dict1.get(elem, 0)
    
    print(number)
```
