```python
with open('input.txt') as fin:
    N, M = map(int, fin.readline().split())
    
    A, B = set(), set()
    
    for i in range(N):
        A.add(int(fin.readline().strip()))
        
    for i in range(M):
        B.add(int(fin.readline().strip()))
        
    arr = [A & B, A - (A & B), B - (A & B)]
        
    for elem in arr:
        print(len(elem))
        print(*sorted(elem))
```
