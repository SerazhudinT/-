```python
with open('input.txt') as fin:
    N = int(fin.readline())
    
    allStudents = set()
    anyStudents = set()
    
    for i in range(N):
        M = int(fin.readline())
        
        now = set(fin.readline().strip() for j in range(M))
        
        allStudents &= now
        anyStudents |= now
        
        if i == 0:
            allStudents = now.copy()
            
    print(len(allStudents))
    print(*allStudents, sep='\n')
    print(len(anyStudents))
    print(*anyStudents, sep='\n')
```
