```python
def extend(rect, d):
    return [rect[0] - d, rect[1] + d, rect[2] - d, rect[3] + d]


def intersect(rect1, rect2):
    return [max(rect1[0], rect2[0]), min(rect1[1], rect2[1]), max(rect1[2], rect2[2]), min(rect1[3], rect2[3])]


with open('input.txt') as fin:
    t, d, n = map(int, fin.readline().split())
    
    rect =  (0, 0, 0, 0)
        
    for _ in range(n):
        navX, navY = map(int, fin.readline().split())
        
        navRect = extend((navX + navY, navX + navY, navX - navY, navX - navY), d)
                    
        rect = intersect(extend(rect, t), navRect)
        
    coords = []
    
    for i in range(rect[0], rect[1] + 1):
        for j in range(rect[2], rect[3] + 1):
            if (i + j) % 2 == 0:
                x = (i + j) // 2
                y = i - x
                coords.append((x, y))
        
    print(len(coords))
    for elem in coords:
        print(*elem)
```
