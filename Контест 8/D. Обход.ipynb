```python
def addNode(tree, node):
    if tree[0] > node:
        if tree[1] == None:
            tree[1] = [node, None, None]
        else:
            addNode(tree[1], node)
    elif tree[0] < node:
        if tree[2] == None:
            tree[2] = [node, None, None]
        else:
            addNode(tree[2], node)


def printTree(tree):
    if tree[1] != None:
        printTree(tree[1])
        
    print(tree[0])

    if tree[2] != None:
        printTree(tree[2])
        

with open('input.txt') as fin:
    lst = list(map(int, fin.readline().split()))
    
    tree = [lst[0], None, None]
    
    for node in lst[1:-1]:
        addNode(tree, node)
        
    printTree(tree)
```
