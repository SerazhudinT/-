```python
def addNode(tree, node):     
    if tree[0] > node:
        if tree[1] is None:
            tree[1] = [node, None, None]
        else:
            addNode(tree[1], node)
    elif tree[0] < node:
        if tree[2] is None:
            tree[2] = [node, None, None]
        else:
            addNode(tree[2], node)

            
def height(tree, n):
    count1 = count2 = n
    
    if tree[1]:
        count1 = height(tree[1], n + 1)
    
    if tree[2]:
        count2 = height(tree[2], n + 1)
        
    return max(count1, count2)
    
            
with open('input.txt') as fin:
    lst = list(map(int, fin.readline().split()))
    
    tree = [lst[0], None, None]
    
    for elem in lst[1:-1]:
        addNode(tree, elem)    

    print(height(tree, 1))
```
