```python
import sys

sys.setrecursionlimit(100000)

def countLevel(dct, parent):
    for child in dct[parent][0]:
        countLevel(dct, child)
        dct[parent][2] += dct[child][2] + 1
        

with open('input.txt') as fin:
    N = int(fin.readline())
    
    dct = {}
    for i in range(N - 1):
        child, parent = fin.readline().split()
        
        if child not in dct:
            dct[child] = [[], [], 0]
            
        if parent not in dct:
            dct[parent] = [[], [], 0]
            
        dct[child][1].append(parent)
        dct[parent][0].append(child)
        
    for first in dct:
        if len(dct[first][1]) == 0:
            break
            
    countLevel(dct, first)
    
    for name in sorted(dct):
        print(name, dct[name][2])
```
