```python
def addNode(tree, node):
    if tree[0] > node:
        if tree[1] == None:
            tree[1] = [node, None, None]
        else:
            addNode(tree[1], node)
    elif tree[0] < node:
        if tree[2] == None:
            tree[2] = [node, None, None]
        else:
            addNode(tree[2], node)              


def leavesHeight(tree, n):
    h1 = h2 = n
    
    if tree[1]:
        h1 = leavesHeight(tree[1], n + 1)
        
    if tree[2]:
        h2 = leavesHeight(tree[2], n + 1)
    
    if abs(h1 - h2) > 1:
        return -1
        
    return max(h1, h2)


with open('input.txt') as fin:
    lst = list(map(int, fin.readline().split()))
    
    tree = [lst[0], None, None]
    
    for node in lst[1:-1]:
        addNode(tree, node)
      
    ans = leavesHeight(tree, 1)
    
    if ans > 0:
        print('YES')
    else:
        print('NO')
```
