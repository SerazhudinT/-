```python
def addNode(tree, node, n):
    if tree[0] > node:
        if tree[1] == None:
            tree[1] = [node, None, None]
            n += 1
        else:
            n = addNode(tree[1], node, n + 1)
    elif tree[0] < node:
        if tree[2] == None:
            tree[2] = [node, None, None]
            n += 1
        else:
            n = addNode(tree[2], node, n + 1)
            
    return n
    
with open('input.txt') as fin:
    lst = list(map(int, fin.readline().split()))
    
    tree = [lst[0], None, None]
    
    st = set([lst[0]])
    height = [1]
    for node in lst[1:-1]:
        if node not in st:
            height.append(addNode(tree, node, 1))
            st.add(node)
    
    print(*height)
```
