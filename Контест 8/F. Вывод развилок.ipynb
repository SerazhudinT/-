```python
def addNode(tree, node):
    if tree[0] > node:
        if tree[1] == None:
            tree[1] = [node, None, None]
        else:
            addNode(tree[1], node)
    elif tree[0] < node:
        if tree[2] == None:
            tree[2] = [node, None, None]
        else:
            addNode(tree[2], node)

            
def printForks(tree):
    if tree[1] != None:
        printForks(tree[1])
        
    if tree[1] and tree[2]:
        print(tree[0])

    if tree[2] != None:
        printForks(tree[2])


with open('input.txt') as fin:
    lst = list(map(int, fin.readline().split()))
    
    tree = [lst[0], None, None]
    
    for node in lst[1:-1]:
        addNode(tree, node)
        
    printForks(tree)
```
