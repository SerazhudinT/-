```python
def addNode(tree, node):
    if tree[0] > node:
        if tree[1] == None:
            tree[1] = [node, None, None]
        else:
            addNode(tree[1], node)
    if tree[0] < node:
        if tree[2] == None:
            tree[2] = [node, None, None]
        else:
            addNode(tree[2], node)
            
            
def searchMax1(tree):
    if tree[2] != None:
        return searchMax1(tree[2])
    else:
        return tree[0]


def searchMax2(tree):
    if tree[2] != None:
        if tree[2][1] or tree[2][2]:
            return searchMax2(tree[2])
        else:
            return tree[0]
    elif tree[1] != None:
        return searchMax1(tree[1])


with open('input.txt') as fin:
    lst = list(map(int, fin.readline().split()))
    
    tree = [lst[0], None, None]
    
    for node in lst[:-1]:
        addNode(tree, node)
    
    print(searchMax2(tree))
```
