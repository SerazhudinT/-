```python
with open('input.txt') as fin:
    n, m = map(int, fin.readline().split())
    
    lst = []
    for i in range(n):
        a, b = map(int, fin.readline().split())
        
        if a > b:
            b, a = a, b
        
        lst.append((a, -1))
        lst.append((b, 1))
    
    lstM = list(map(int, fin.readline().split()))
    for i in range(m):
        lst.append((lstM[i], 0, i))
    
    lst.sort()
    
    ans = []
    numA = numB = 0
    for elem in lst:
        if elem[1] == -1:
            numA += 1
        elif elem[1] == 1:
            numB += 1
        elif elem[1] == 0:
            ans.append((elem[2], numA - numB))   
       
    print(*map(lambda x: x[1], sorted(ans)))
```
