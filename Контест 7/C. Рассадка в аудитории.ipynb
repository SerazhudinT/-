```python
with open('input.txt')as fin:
    N, D = map(int, fin.readline().split())
    lst = list(map(int, fin.readline().split()))
    
    for i in range(N):
        lst[i] = (lst[i], i)
    
    lst.sort()
    
    l = 1
    r = N
    while l < r:
        m = (l + r) // 2
        
        i = 0
        while i + m < N and lst[i + m][0] - lst[i][0] > D:
            i += 1
        
        if i == N - m:
            r = m
        else:
            l = m + 1
    
    tickets = []
    for i in range(N):
        tickets.append((i % l + 1, lst[i][1]))
        
    print(l)  
    print(*map(lambda x: x[0], sorted(tickets, key=lambda y: y[1])))
```
