```python
with open('input.txt') as fin:
    N, M = map(int, fin.readline().split())
    
    lst = []
    for i in range(M):
        a, b = map(int, fin.readline().split())
        
        lst.append((a, -1))
        lst.append((b, 1))

    lst.sort()
    
    count = N
    a = b = -1
    numA = numB = 0
    for elem in lst:
        if elem[1] == -1:
            numA += 1
        elif elem[1] == 1:
            numB += 1
        
        if a == -1:
            a = elem[0]
        
        if numA == numB:
            b = elem[0]
            count -= b - a + 1
            
            a = -1
    
    print(count)
```
