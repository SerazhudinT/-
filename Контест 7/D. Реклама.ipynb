```python
with open('input.txt') as fin:
    N = int(fin.readline())
    
    notSortedLst = []
    for i in range(N):
        a, b = map(int, fin.readline().split())
        
        if b - a >= 5:
            notSortedLst.append([a, -1, i])
            notSortedLst.append([b, 1, i])
        
    lst = sorted(notSortedLst)
    
    maxLevel = 0
    maxStart = 0
    maxIndex = 0
    maxPeople = set()
    
    curPeople = set()
    numA = numB = 0
    for elem in lst:
        if elem[1] == -1:
            numA += 1
            curPeople.add(elem[2])
        elif elem[1] == 1:
            numB += 1
            curPeople.discard(elem[2])
        
        if maxLevel < numA - numB:
            maxLevel = numA - numB
            maxStart = elem[0]
            maxIndex = elem[2]
            maxPeople = curPeople.copy()
            
    if notSortedLst[maxIndex + 1][0] - notSortedLst[maxIndex][0] < 10:
        notSortedLst.remove(notSortedLst[maxIndex])
        notSortedLst.remove(notSortedLst[maxIndex])
    else:
        notSortedLst[maxIndex][0] += 5
  

    lst = sorted(notSortedLst)
    
    maxLevel2 = 0
    maxStart2 = 0
    maxPeople2 = set()
    
    curPeople = set()
    numA = numB = 0
    for elem in lst:
        if elem[1] == -1:
            numA += 1
            curPeople.add(elem[2])
        elif elem[1] == 1:
            numB += 1
            curPeople.discard(elem[2])
        
        if len(maxPeople2 - maxPeople) < len(curPeople - maxPeople):
            maxLevel2 = numA - numB
            maxStart2 = elem[0]
            maxPeople2 = curPeople - maxPeople
    
    print(maxLevel2, maxStart2, maxPeople2)
            
    print(maxLevel, maxStart, maxPeople)
```
