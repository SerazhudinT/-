```python
with open('input.txt') as fin:
    N = int(fin.readline())
    
    maxLen = curLen = 0
    for i in range(N):
        elem = int(fin.readline())
        
        if elem == 1:
            curLen += 1
            maxLen = max(maxLen, curLen)
        else:
            curLen = 0
    
    print(maxLen)
```
