```python
def isAnagramm(s1, s2):
    if len(s1) != len(s2):
        return 0
    
    dct1 = {}
    dct2 = {}
    for i in range(len(s1)):
        dct1[s1[i]] = dct1.get(s1[i], 0) + 1
        dct2[s2[i]] = dct2.get(s2[i], 0) + 1
    
    return int(dct1 == dct2)

with open('input.txt') as fin:
    s1 = fin.readline().strip()
    s2 = fin.readline().strip()
    
    print(isAnagramm(s1, s2))
```
