```python
def makeRoutes(cities, maxDist, A, B, routesLen, curRoute):
    curRoute.append(A)
    
    if B in cities[A]:
        routesLen.append(len(curRoute))
    else:
        for city in cities[A]:
            if city not in curRoute:
                makeRoutes(cities, maxDist, city, B, routesLen, curRoute)
        
    curRoute.pop()

    
with open('input.txt') as fin:
    N = int(fin.readline())
    
    cities = []
    for i in range(N):
        x, y = map(int, fin.readline().split())
        
        cities.append((x, y))
        
    maxDist = int(fin.readline())
    
    A, B = map(int, fin.readline().split())
    A = cities[A - 1]
    B = cities[B - 1]
    
    avCities = {}
    for i in range(N):
        for j in range(i + 1, N):
            if abs(cities[i][0] - cities[j][0]) + abs(cities[i][1] - cities[j][1]) <= maxDist:
                if cities[i] not in avCities:
                    avCities[cities[i]] = []
                    
                if cities[j] not in avCities:
                    avCities[cities[j]] = []
                    
                avCities[cities[i]].append(cities[j])
                avCities[cities[j]].append(cities[i])
        
    routesLen = []
    curRoute = []
    
    if len(avCities) != 0:
        makeRoutes(avCities, maxDist, A, B, routesLen, curRoute)
    
    print(min(routesLen, default=-1))
```
