```python
with open('input.txt') as fin:
    a, b, c = map(int, fin.read().split())
    
    l = 0
    r = a + b + c
    
    while l < r:
        m = (l + r) // 2
        
        if 2 * (a * 2 + b * 3 + c * 4 + m * 5) >= 7 * (a + b + c + m):
            r = m
        else:
            l = m + 1
            
    print(l)
```
