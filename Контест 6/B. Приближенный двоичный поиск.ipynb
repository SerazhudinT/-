```python
with open('input.txt') as fin:
    N, K = map(int, fin.readline().split())
    lst = list(map(int, fin.readline().split()))
    lstK = list(map(int, fin.readline().split()))
    
    for elem in lstK:
        l = 0
        r = N - 1
        
        while l < r:
            m = (l + r) // 2
            
            if elem < lst[m]:
                r = m
            else:
                l = m + 1
        
        if abs(elem - lst[l - 1]) <= abs(elem - lst[l]):
            print(lst[l - 1])
        else:
            print(lst[l])
```
