```python
with open('input.txt') as fin:
    N, x, y = map(int, fin.read().split())
    
    if x > y:
        x, y = y, x
        
    N -= 1
        
    l = 0
    r = N
    
    while l < r:
        m = (l + r) // 2
        
        if (N - m) * y <= m * x:
            r = m
        else:
            l = m + 1
    
    print(x + min(max(l * x, (N - l) * y), max((l - 1) * x, (N - (l - 1)) * y)))
```
