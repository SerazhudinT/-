```python
with open('input.txt') as fin:
    N, R, C = map(int, fin.readline().split())
    height = sorted(map(int, fin.read().split()))
    
    l = 0
    r = height[-1] - height[0]
    while l < r:
        m = (l + r) // 2
        
        i = 0
        count = 0
        while i + C - 1 < N:
            if height[i + C - 1] - height[i] <= m:
                count += 1
                i += C
            else:
                i += 1
                
        if count >= R:
            r = m
        else:
            l = m + 1
            
    print(l)
```
