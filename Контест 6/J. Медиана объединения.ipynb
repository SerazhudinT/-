```python
with open('input.txt') as fin:
    N, L = map(int, fin.readline().split())
    
    seqs = []
    for seq in fin:
        seqs.append(list(map(int, seq.split())))
        
    for i in range(N):
        for j in range(i + 1, N):
            curI = seqs[i]
            curJ = seqs[j]
            
            k = t = 0
            while k + t < L - 1:
                if curI[k] <= curJ[t]:
                    k += 1
                else:
                    t += 1
                    
            print(min(curI[k], curJ[t]))
```
