```python
with open('input.txt') as fin:
    w, h, n = map(int, fin.readline().split())
    
    l = 0
    r = n * max(w, h)
    
    while l < r:
        m = (l + r) // 2
        
        if (m // w) * (m // h) >= n:
            r = m
        else:
            l = m + 1
            
    print(l)
```
