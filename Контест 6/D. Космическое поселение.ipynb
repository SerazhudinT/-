```python
def binSearch(n, a, b, w, h):
    l = 0
    r = w
    
    while l < r:
        d = (l + r + 1) // 2
        
        if (w // (a + 2 * d)) * (h // (b + 2 * d)) >= n:
            l = d
        else:
            r = d - 1
            
    return l

with open('input.txt') as fin:
    n, a, b, w, h = map(int, fin.readline().split())
      
    print(max(binSearch(n, a, b, w, h), binSearch(n, b, a, w, h)))
```
