```python
with open('input.txt') as fin:
    N, K = map(int, fin.readline().split())
    lstN = list(map(int, fin.readline().split()))
    lstK = list(map(int, fin.readline().split()))    

    for elem in lstK:
        l = 0
        r = N - 1
        
        while l < r:
            m = (l + r) // 2
            
            if elem <= lstN[m]:
                r = m
            else:
                l = m + 1
        
        if elem == lstN[l]:
            print('YES')
        else:
            print('NO')
```
