```python
with open('input.txt') as fin:
    n, m, t = map(int, fin.read().split())
    
    l = 0
    r = n
    
    while l < r:
        d = (l + r + 1) // 2
        
        a = n - 2 * d
        b = m - 2 * d
        
        if a > 0 and b > 0 and n * m - a * b <= t:
            l = d
        else:
            r = d - 1
            
    print(l)
```
