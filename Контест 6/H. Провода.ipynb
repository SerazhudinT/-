```python
with open('input.txt') as fin:
    N, K = map(int, fin.readline().split())
    lst = list(map(int, fin.read().split()))
    
    l = 1
    r = max(lst)
    
    while l < r:
        m = (l + r + 1) // 2
        
        count = 0
        for elem in lst:
            count += elem // m
            
        if count >= K:
            l = m
        else:
            r = m - 1
    
    if sum(lst) < K:
        l = 0
        
    print(l)
```
