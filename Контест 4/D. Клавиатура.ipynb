```python
with open('input.txt') as fin:    
    n = int(fin.readline())
    lst = list(map(int, fin.readline().split()))
    
    dct = {i + 1: lst[i] for i in range(n)}
    
    k = int(fin.readline())

    for keys in map(int, fin.readline().split()):
        dct[keys] -= 1
    
    for keys in sorted(dct):
        if dct[keys] < 0:
            print('YES')
        else:
            print('NO')
```
