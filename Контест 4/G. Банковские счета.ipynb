```python
with open('input.txt') as fin:    
    dct = {}
    
    for string in fin:
        string = string.split()
        
        if string[0] == 'DEPOSIT':
            dct[string[1]] = dct.get(string[1], 0) + int(string[2])
        elif string[0] == 'WITHDRAW':
            dct[string[1]] = dct.get(string[1], 0) - int(string[2])
        elif string[0] == 'BALANCE':
            if dct.get(string[1]) is not None:
                print(dct[string[1]])
            else:
                print('ERROR')
        elif string[0] == 'TRANSFER':
            dct[string[1]] = dct.get(string[1], 0) - int(string[3])
            dct[string[2]] = dct.get(string[2], 0) + int(string[3])
        elif string[0] == 'INCOME':
            for name in dct:
                if dct[name] > 0:
                    dct[name] += int(dct[name] * int(string[1]) / 100)
```
