```python
with open('input.txt') as fin:    
    dct = {}
    
    for string in fin:
        name, product, number = string.split()
        
        nameProd = name + ' ' + product
        
        dct[nameProd] = dct.get(nameProd, 0) + int(number)
        
    prevName = ''
    
    for nameProd in sorted(dct):
        name, product = nameProd.split()
        
        if name != prevName:
            print(name + ':')
            prevName = name
        
        print(product, dct[nameProd])
```
