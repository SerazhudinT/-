```python
def dictIsEqual(dct, curDct):
    for char in dct:
        if dct[char] != curDct[char]:
            return False
        
    return True
        
    
with open('input.txt') as fin:
    g, n = map(int, fin.readline().split())
    
    W = fin.readline().strip()
    S = fin.readline().strip()
    
    dct = {}
    curDct = {}

    for char in W:
        dct[char] = dct.get(char, 0) + 1
        
    for char in set(S):
        curDct[char] = curDct.get(char, 0)

    number = 0
    for i in tqdm(range(n)):
        curDct[S[i]] += 1
        
        if i >= g:
            curDct[S[i - g]] -= 1
            
        if dct.get(S[i]) is not None and dictIsEqual(dct, curDct):
            number += 1
                    
    print(number)
```
