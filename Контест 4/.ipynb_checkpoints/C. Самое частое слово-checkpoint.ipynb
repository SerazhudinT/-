```python
with open('input.txt') as fin:
    dct = {}
    
    for string in fin:
        string = string.split()
        
        for word in string:
            dct[word] = dct.get(word, 0) + 1
            
    maxPeriod = max(dct.values())
    
    maxWord = sorted(word for word in dct if dct[word] == maxPeriod)[0]
    
    print(maxWord)
```
