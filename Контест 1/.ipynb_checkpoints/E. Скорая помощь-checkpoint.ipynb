```python
import math

def defK_N(K2, P2, N2, K_N):
     return K2 - (P2 - 1) * M * K_N - (N2 - 1) * K_N

def defP_N(K1, M, K_N):
    P1 = math.ceil(K1 / (M * K_N))
    N1 = math.ceil((K1 - (P1 - 1) * M * K_N) / K_N)
    
    return P1, N1

K1, M, K2, P2, N2 = map(int, input().split())
P1 = N1 = 0
    
P1_N1 = [defP_N(K1, M, i) for i in range(10**4) if 1 <= defK_N(K2, P2, N2, i) <= i]

if len(P1_N1) == 0 or N2 > M:
    P1 = N1 = -1
else:
    P1_N1 = list(map(lambda x: list(set(x)), zip(*P1_N1)))

    if len(P1_N1[0]) == 1:
        P1 = P1_N1[0][0]

    if len(P1_N1[1]) == 1:
        N1 = P1_N1[1][0]
            
print(P1, N1)
```
