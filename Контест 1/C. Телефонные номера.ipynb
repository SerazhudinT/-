```python
numbers = [input() for i in range(4)]

for i in range(4):
    numbers[i] = numbers[i].strip('+')
    numbers[i] = ''.join(numbers[i].split('-'))
    numbers[i] = ''.join(numbers[i].split(')'))
    numbers[i] = ''.join(numbers[i].split('('))
        
    if len(numbers[i]) == 11:
        numbers[i] = numbers[i][1:]
        
        if numbers[i][:3] == '495':
            numbers[i] = numbers[i][3:]
    
    if i > 0:
        if numbers[i] == numbers[0]:
            print('YES')
        else:
            print('NO')
```
