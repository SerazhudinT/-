```python
with open('input.txt') as fin:
    n, k = map(int, fin.readline().split())
    x = map(int, fin.readline().split())
    
    dct = {}
    for elem in x:
        if dct.get(elem, 0) < 3:
            dct[elem] = dct.get(elem, 0) + 1
            
    dctKeys = sorted(dct.keys())
    
    j = 0
    count = 0
    dctLen = len(dct)
    repLst = 4 * [0]
    
    for i in range(dctLen):
        while j < dctLen and k * dctKeys[i] >= dctKeys[j]:
            repLst[dct[dctKeys[j]]] += 1
            j += 1
        
        repLst[dct[dctKeys[i]]] -= 1
        
        uniqCnt = j - i

        if dct[dctKeys[i]] >= 1:
            count += 6 * (uniqCnt - 1) * (uniqCnt - 2) + 3 * (repLst[2] + repLst[3])

        if dct[dctKeys[i]] >= 2:
            count += 3 * (uniqCnt - 1)

        if dct[dctKeys[i]] == 3:
            count += 1

    print(count)
```
