```python
with open('input.txt') as fin:
    N, K = map(int, fin.readline().split())
    lstN = list(map(int, fin.readline().split()))   
    
    count = 0
    alongSum = 0
    j = 0
    for i in range(N):
        while j < N and alongSum < K:
            alongSum += lstN[j]
            j += 1
            
        if alongSum == K:
            count += 1
            
        alongSum -= lstN[i]
        
    print(count)
```
