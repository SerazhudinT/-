```python
with open('input.txt') as fin:
    K = int(fin.readline())
    s = fin.readline().strip()
    
    j = K
    count = 0
    lenS = len(s)
    for i in range(lenS):
        while j < lenS and s[j] == s[i + (j - i) % K]:
            j += 1
            
        count += max(j - i - K, 0)
        
    print(count)
```
