```python
with open('input.txt') as fin:
    N = int(fin.readline())
    
    hPlus = [0]
    hMinus = [0]
    x0, y0 = map(int, fin.readline().split())
    for _ in range(N - 1):
        x, y = map(int, fin.readline().split())
        
        if y - y0 > 0:
            hPlus.append(y - y0 + hPlus[-1])
            hMinus.append(hMinus[-1])
        else:
            hPlus.append(hPlus[-1])
            hMinus.append(y0 - y + hMinus[-1])
            
        y0 = y
    
    M = int(fin.readline())
    
    for _ in range(M):
        s, f = map(int, fin.readline().split())
        
        s, f = s - 1, f - 1
        
        if f - s > 0:
            print(hPlus[f] - hPlus[s])
        else:
            print(hMinus[s] - hMinus[f])
```
