```python
with open('input.txt') as fin:
    n, r = map(int, fin.readline().split())
    dist = list(map(int, fin.readline().split()))
    
    count = 0
    j = 1
    for i in range(n):
        while j < n and dist[j] - dist[i] <= r:
            j += 1
        
        if j < n and dist[j] - dist[i] > r:
            count += n - j
    
    print(count)
```
