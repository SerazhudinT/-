```python
with open('input.txt') as fin:
    N, K = map(int, fin.readline().split())
    row = list(map(int, fin.readline().split()))
    
    trees = (K + 1) * [0]
    cntTrees = 0
    
    segment = 1, N
    j = 0
    for i in range(N):
        while j < N and cntTrees < K:
            trees[row[j]] += 1
            
            if trees[row[j]] == 1:
                cntTrees += 1
            
            j += 1
        
        if cntTrees == K and segment[1] - segment[0] > j - i - 1:
            segment = i + 1, j
        
        trees[row[i]] -= 1
        
        if trees[row[i]] == 0:
            cntTrees -= 1
                
    print(*segment)  
```
