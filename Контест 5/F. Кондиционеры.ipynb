```python
with open('input.txt') as fin:
    n = int(fin.readline())
    classes = list(map(int, fin.readline().split()))
    
    powerPrice = 1001 * [1001]
    
    m = int(fin.readline())
    for i in range(m):
        power, price = map(int, fin.readline().split())
        
        powerPrice[power] = min(powerPrice[power], price)
        
    for i in range(999, 0, -1):
        powerPrice[i] = min(powerPrice[i], powerPrice[i + 1])
        
    cost = 0
    for elem in classes:
        cost += powerPrice[elem]
    
    print(cost)
```
