```python
with open('input.txt') as fin:
    n, k = map(int, fin.readline().split())
    s = fin.readline().strip()
    
    dct = {}
    
    j = 0
    maxSubStr = 1, 1
    for i in range(n):
        while j < n and dct.get(s[j], 0) < k:
            dct[s[j]] = dct.get(s[j], 0) + 1
            j += 1
            
        if maxSubStr[0] < j - i:
            maxSubStr = j - i, i + 1
            
        dct[s[i]] -= 1
    
    print(*maxSubStr)
```
